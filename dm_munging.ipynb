```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
```

```python
map_df = pd.read_csv("MDL_adoptive_metadata.txt", sep='\t', index_col='#SampleID')

    
```

```python
dm_df = pd.read_csv("weighted_unifrac_dm.txt", sep='\t', index_col=0)
dm_df[:5]
```

```python
#stack the series and filter out the duplicates 
dm_series = dm_df.stack()
seen = []

for id_pair in dm_series.index.tolist():
    if id_pair[::-1] in seen:
        continue
    seen.append(id_pair)

dm_series = dm_series[seen]
```

```python
stack_df = DataFrame(dm_series)
stack_df.columns = ['dist']
stack_df[:5]
```

```python
stack_df['Person'] = np.nan
stack_df['Person']['FN4Dad.Mouth.TP2'] = map_df['Person']['FN4Dad.Mouth.TP2']
stack_df.ix[:5]
```

```python
def concat(x):
    return x + map_df.loc[x.name[0]]
    
%timeit stack_df.apply(concat, axis=1)
```

```python
for col in map_df.columns:
    stack_df[col] = np.nan
    for i in stack_df.index:
        stack_df[col][i[0]] = map_df[col][i[0]]
```

```python
lookup = {}
bool_list = []

def check_duplicate(x):
    keya = x.name[0]
    keyb = x.name[1]
    
    def check_a():
        if keya in lookup:
            if keyb in lookup[keya]:
                return False
            else:
                return check_b()
        else:
            return check_b()
        
    def check_b():
        if keyb in lookup:
            if keya in lookup[keyb]:
                return False
            else:
                return add()
        else:
            return add()
    
    def add():
        if keya not in lookup:
            lookup[keya] = {}
        lookup[keya][keyb] = True
        return True
        
    
    bool_list.append(check_a())
    
stack_df.apply(check_duplicate, axis=1)
new_df = stack_df[bool_list]
len(new_df)
    
```

```python
# list of categories to keep BodySite, Person, adopt_status, Birthplace, Family_Number, Time_Point, child, parent
categories = ['BodySite', 'Person', 'adopt_status', 'Birthplace', 'Family_Number', 'Time_Point', 'Parent', 'dist']
all_cats = stack_df.columns.tolist()
cats_to_discard = [x for x in all_cats if x not in categories]
stack_df = stack_df.drop(cats_to_discard, axis=1)
stack_df
```

```python
stack_df.to_csv('filtered_table.txt', sep='\t')
```
